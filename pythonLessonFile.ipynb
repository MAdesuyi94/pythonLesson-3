## if statement

### - For example, if age equals 16, then you can drive

- For example, ```if age >= 16:```
### The code block within (below) the if statement is indented by pressing the tab (or 4 spaces)
- For example, ```print("You can drive")```


## Boolean Expression
### A boolean expression is an expression that evaluates to either True or False.
### Example: a > b
- If a is in fact greater than b, then this statement will return `True`
- If a is NOT greater than b, then this statement will return `False`

## Relational Operator
### A relational operator is used to compare two values. It determins whether a relationship exists between two values
### Relational Operators
- `<` - Less Than
- `>` - Greater Than
- `<=` - Less Than or Equal To
- `>=` - Greater Than or Equal To
- `==` - Equal To
- `!=` - Not Equal To

### Boolean Expressions with Relational Operators
- `x < y` - x is less than y
- `x > y` - x is greater than y
- `x <= y` - x is less than or equal to y
- `x >= y` - x is greater than or equal to y
- `x == y` - x is equal to y
- `x != y` - x is not equal to y

#### Note: for "equals to", you have to do double equal sign `==` NOT single equal sign `=`. Single equal sign is for variable assignment.

## If-Else Statements

### An if-else statement is a control flow that allows one code block to execute if the condition is true, and another code block to execute if the condition is false.
### An if-else statement starts with the keyword "if" followed by a condition then a colon (:). The code block within (below) the if statement is indented by pressing the tab (or 4 spaces). After the if code block, you write the keyword "else" followed by a colon (:). The code block within (below) the else statement is also indented by pressing the tab (or 4 spaces).
```python
age = 16
if age >= 16:
    print("You can drive")
else:
    print("You cannot drive")
```

## If-Elif-Else Statements
### An if-elif-else statement is a control flow that allows one code block to execute if the first condition is true, another code block to execute if the second condition is true, and another code block to execute if neither condition is true.
### - An if-elif-else statement starts with the keyword "if" followed by a condition then a colon (:). The code block within (below) the if statement is indented by pressing the tab (or 4 spaces). 
### - After the if code block, you write the keyword "elif" followed by a second condition then a colon (:). The code block within (below) the elif statement is also indented by pressing the tab (or 4 spaces).
### - After the elif code block, you write the keyword "else" followed by a colon (:). The code block within (below) the else statement is also indented by pressing the tab (or 4 spaces).
```python
age = 16
if age >= 18:
    print("You can vote and drive")
elif age >= 16:
    print("You can drive")
else:
    print("You cannot vote or drive")
```

## `and` operator
### The `and` operator is a logical operator that combines two boolean expressions and returns True only if both expressions are True.
```python
age = 20
if age >= 16 and age < 18:
    print("You can drive but cannot vote")
else:
    print("You can either vote or cannot drive")
```

## `or` operator
### The `or` operator is a logical operator that combines two boolean expressions and returns True if at least one of the expressions is True.
```python
age = 20
if age >= 16 or age >= 18:
    print("You can either vote or drive")
else:
    print("You cannot vote or drive")
```

## `not` operator
### The `not` operator is a logical operator that negates a boolean expression. It returns True if the expression is False, and returns False if the expression is True.
```python
age = 20
if not age < 16:
    print("You can drive")
else:
    print("You cannot drive")
```
